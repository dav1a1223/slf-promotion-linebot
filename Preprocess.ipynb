```python
from html.parser import HTMLParser
import json
import os
from os import listdir
import datetime
import operator
import pdb
```

```python
def load_json(file_name):
    with open(file_name) as f:
        data = json.load(f)
    messages = data['messages']
    new_st = [{k:(v.encode('iso-8859-1').decode('utf-8') if type(v) == str else v) for k, v in st.items()} for st in messages]
    new_st.sort(key=operator.itemgetter('timestamp_ms'))
    return new_st
```

```python
def preprocess(sorted_json):
    pre_time = 0
    all_docs = []
    doc = {}
    name = ''
    name_list = []
    content = ''
    QA_pair = {'from_doc':0,'pair':[]}
    QA_pairs = []
    doc_id = 0
    global owner_name
    for st in sorted_json:
        if('content' in st and '你傳送了 1 張貼圖。' in st['content']):
            continue
        if('content' in st and '錯過了你的來電' in st['content']):
            continue
        if(st['timestamp_ms']-pre_time > 1500):
            if(not doc == {}):
                doc['contents'].append({'sender_name':name,'content':content})
                doc['end_time'] = pre_time
                all_docs.append(doc)
    #             content = ''
            doc = {'id':doc_id,'contents':[],'start_time':st['timestamp_ms']}
            doc_id += 1
    #         name = ''
        pre_time = st['timestamp_ms']
        if('content' in st):
            if(st['sender_name'] == name):
                content += '\n' + st['content']
            else:
    #             content += 
                if(name == owner_name):
                    if(QA_pair['pair'] != []):
                        QA_pair['pair'].append(content)
                        QA_pairs.append(QA_pair)
                        QA_pair = {'from_doc':0,'pair':[]}
                else:
                    QA_pair['from_doc'] = doc_id-1
                    QA_pair['pair'] = [content]
    #             if(doc['contents'] != []):
                doc['contents'].append({'sender_name':name,'content':content})
                content = st['content']
                name = st['sender_name']
    return all_docs, QA_pairs
```

## 記得改 data_path, output_path, owner_name
## QA_pairs: owner 有回答的問題跟答案，會說這個 pair 從第幾個 document 來，list 的第一個是 Q 第二個是 A
## all_docs: 把所有對話記錄根據 timestamp 差 1500 以上就切開切成一堆document，可以拿來做topic model

```python
data_path = 'messages/inbox' #json檔案的位置
output_path = 'processed_messages/' #開一個資料夾放做好的檔案
QA_pair_size_threshold = 20 #要有幾個 QA pair 才做成檔案
all_folder_names = listdir(data_path)
for folder_name in all_folder_names:
    print(folder_name)
    json_file = data_path + '/' + folder_name + '/message.json'
    owner_name = '賴冠廷' #你的名字
    try:
        sorted_message = load_json(json_file)
        all_docs, QA_pairs = preprocess(sorted_message)
        if(len(QA_pairs) > QA_pair_size_threshold):
            with open(output_path+folder_name+'.docs','w') as f1:
                json.dump(all_docs,f1)
            with open(output_path+folder_name+'.pairs','w') as f1:
                json.dump(QA_pairs,f1)
    except:
        continue
```
